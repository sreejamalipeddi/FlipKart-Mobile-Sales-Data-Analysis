**Project Title** : **FlipKart Mobile Sales Data Analysis**

**Purpose**:
This study provides an analysis of Flipkart mobile sales data to uncover key insights and trends in the market. Our primary goals are to identify the best-selling mobile brands and models, analyze pricing strategies, and understand consumer preferences.

**Step 6 : Data Visualization**

```python
# Assigning data to df variable
df = pd.DataFrame(dataset)

# Creating a pie chart for Brand
df['Brand'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Brand',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating a pie chart for Memory
df['Memory'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Memory',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating a pie chart for Storage
df['Storage'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('storage',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating bar chart for top 10 Brands sold
plt.figure(figsize=(6,4))
df['Brand'].value_counts().head(10).plot(kind='bar',color='#FF0067',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Brands',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Brands sold',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating bar chart for top 10 Model sold
plt.figure(figsize=(6,4))
df['Model'].value_counts().head(10).plot(kind='bar',color='#1f77b4',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Models',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Models sold',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating histogram of selling price distribution
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(df['Selling Price'],color='#ff7f0e',edgecolor='black')
plt.title('Selling Price Distribution')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')

# Creating histogram of Brand distribution
plt.subplot(2, 2, 2)
plt.hist(df['Brand'], color='#ff7f0e',edgecolor='black')
plt.title('Based on Brand')
plt.xlabel('Brand')
plt.ylabel('Frequency')

# Creating histogram of Original Price distribution
plt.subplot(2, 2, 3)
plt.hist(df['Original Price'],color='#ff7f0e', edgecolor='black')
plt.title('based on original Price')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Creating histogram of Rating distribution
plt.subplot(2, 2, 4)
plt.hist(df['Rating'], color='#ff7f0e',edgecolor='black')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Creating bar chart of Mobiles sold based on Brand
plt.figure(figsize=(8,4))
df['Brand'].value_counts().sort_index().plot(kind='bar',color='#2ca02c',edgecolor='black',width=0.8)
plt.title('Count of Mobiles by Brands',weight='bold',fontsize=14)
plt.xlabel('Brands',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
```

```python
# Creating bar chart of number of mobiles sold based on memory
plt.figure(figsize=(6,4))
df['Memory'].value_counts().plot(kind='bar',color='#f34089',edgecolor='black',width=0.8)
plt.title('Count of Mobiles by Memory',weight='bold',fontsize=14)
plt.xlabel('Memory',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.show()
```

```python
# Creating bar chart of number of mobiles sold based on rating
plt.figure(figsize=(6,4))
df['Rating'].value_counts().plot(kind='bar',color='#8749',edgecolor='black',width=0.8)
plt.title('Count of Mobiles by Rating',weight='bold',fontsize=14)
plt.xlabel('Rating',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.show()
```

```python
# Creating line chart to know trend of selling price over Brands
plt.figure(figsize=(8,4))
df.groupby('Brand')['Selling Price'].mean().plot(marker='o')
plt.xlabel('Brand',weight='bold',fontsize=14)
plt.ylabel('Average Selling Price',weight='bold',fontsize=14)
plt.title('Trend of Selling Price over Brands',weight='bold',fontsize=14)
plt.show()
```

```python
# Creating line chart to know trend of selling price over Rating
plt.figure(figsize=(8,4))
df.groupby('Rating')['Selling Price'].mean().plot(marker='o')
plt.xlabel('Rating',weight='bold',fontsize=14)
plt.ylabel('Average Selling Price',weight='bold',fontsize=14)
plt.title('Trend of Selling Price over Rating',weight='bold',fontsize=14)
plt.show()
```

**Conclusion**


The the Flipkart sales data analysis on mobile phones reveals key insights based on brand, model, rating, memory, and storage.


*   High-rated models tend to have larger memory and storage capacities, reflecting consumer preferences for devices with better performance and storage capabilities.
*   Popular brands such as Samsung, Apple, and Xiaomi dominate sales, with models like Samsung Galaxy and iPhone consistently receiving high ratings.


*  Models offering larger memory and storage capacities, typically 8GB RAM and 128GB storage or more, are more popular
*   This analysis underscores the importance of brand reputation, model specifications, and consumer ratings in driving mobile phone sales on Flipkart.
